```python
import json
import pandas as pd
```

```python
INPUT_PATH = 'data/input/new_wordsjson.json'
```

```python
f = open(INPUT_PATH)
database = json.load(f)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# The information that we want are found in cards
pd.json_normalize(database).T
```

```python
pd.json_normalize(database['cards']).head()
```

```python
df_cards = pd.json_normalize(database['cards'])
```

```python
# I only want name and description
df_cards[df_cards['id']== '582da57839884f8b5da9aadd'].T
```

```python
name_list = list(df_cards['name'].values)
desc_list = list(df_cards['desc'].values)
```

```python
df_name_desc = pd.DataFrame( desc_list, name_list, columns=['meaning'])
```

```python
df_name_desc = pd.concat((pd.Series(name_list),pd.Series(desc_list)),axis=1).rename(columns={0:'Word',1:'Desc'})
```

```python
from datetime import datetime

now = datetime.now().strftime("%m%d%Y%H%M%S")
```

```python
OUTPUT_PATH = 'data/output/'
```

```python
df_name_desc.to_csv(OUTPUT_PATH+now+'.csv',index=False, header=False)
```
